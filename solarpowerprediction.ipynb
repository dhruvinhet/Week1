```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['temperature_2_m_above_gnd'], y=df['generated_power_kw'], alpha=0.5, color='blue')
plt.xlabel('Temperature (°C)')
plt.ylabel('Generated Power (kW)')
plt.title('Power Generation vs. Temperature')
plt.show()
```

```python
# Train ARIMA Model
model = ARIMA(df['generated_power_kw'], order=(5,1,0))
model_fit = model.fit()

# Forecast Next 5 Days
forecast = model_fit.forecast(steps=5)
print(forecast)
```
